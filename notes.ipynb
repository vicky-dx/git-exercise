***Merging branches***

once we have created a new branch and made changes to it, we can merge it back to the master branch

main branch is ground truth, we want to merge changes from feature branch to main branch

source: feature branch
destination: main branch

to merge branches, we need to be in the destination branch / main branch

git switch main
git merge <source>

if we are in different branch and want to merge, we can use
`git merge <source> main` (source branch, destination branch)


types of merge:
1. fast forward merge
2. 3 way merge

fast forward merge:
when there are no changes in the destination branch since the source branch was created, git simply moves the pointer of the destination branch to the source branch without creating a new commit

3 way merge:
when there are changes in the destination branch since the source branch was created, git creates a new commit that combines the changes from both branches and moves the pointer of the destination branch to the new commit 

***Merge confilcts***

when git is unable to automatically merge the changes from two branches, it creates a merge conflict. A merge conflict occurs when the changes made in the source branch and the destination branch overlap or conflict with each other. When a merge conflict occurs, git stops the merge process and asks you to resolve the conflict manually.
